```python
# ================================
# 1. IMPORTS + CONNECTION
# ================================
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:Dana.2004@localhost/sakila")

# ================================
# 2. DIMENSIONS
# ================================

# DIM_CUSTOMER
dim_customer = pd.read_sql("""
SELECT customer_id,
CONCAT(first_name,' ',last_name) AS full_name
FROM customer
""", engine)
dim_customer['customer_key'] = range(1, len(dim_customer)+1)

# DIM_FILM
dim_film = pd.read_sql("""
SELECT film_id, title, rental_duration
FROM film
""", engine)
dim_film['film_key'] = range(1, len(dim_film)+1)

# DIM_STORE
dim_store = pd.read_sql("SELECT store_id FROM store", engine)
dim_store['store_key'] = range(1, len(dim_store)+1)

#  DIM_STAFF 
dim_staff = pd.read_sql("""
SELECT staff_id,
CONCAT(first_name,' ',last_name) AS full_name
FROM staff
""", engine)
dim_staff['staff_key'] = range(1, len(dim_staff)+1)

# DIM_DATE
dates = pd.read_sql("""
SELECT rental_date AS d FROM rental
UNION
SELECT payment_date FROM payment
""", engine)
dates['d'] = pd.to_datetime(dates['d'])
dates = dates.dropna().drop_duplicates()

dim_date = pd.DataFrame()
dim_date['date_key'] = dates['d'].dt.strftime('%Y%m%d').astype(int)
dim_date['year'] = dates['d'].dt.year
dim_date['month'] = dates['d'].dt.month

# ================================
# 3. FACT TABLES
# ================================

# -------- FACT_RENTAL --------
fact_rental = pd.read_sql("""
SELECT r.rental_date, r.return_date,
r.customer_id,
r.staff_id,
i.film_id,
i.store_id
FROM rental r
JOIN inventory i ON r.inventory_id=i.inventory_id
""", engine)

fact_rental['rental_date'] = pd.to_datetime(fact_rental['rental_date'])
fact_rental['return_date'] = pd.to_datetime(fact_rental['return_date'])

# معالجة null
fact_rental['return_date'] = fact_rental['return_date'].fillna(fact_rental['rental_date'])

# duration
fact_rental['actual_duration'] = (
    fact_rental['return_date'] - fact_rental['rental_date']
).dt.days

# merge film duration
fact_rental = fact_rental.merge(
    dim_film[['film_id','rental_duration']],
    on='film_id',
    how='left'
)

fact_rental['rental_duration'] = fact_rental['rental_duration'].fillna(0)

# late
fact_rental['is_late'] = fact_rental['actual_duration'] > fact_rental['rental_duration']

#  date_key
fact_rental['date_key'] = fact_rental['rental_date'].dt.strftime('%Y%m%d').astype(int)

#  surrogate keys
fact_rental = fact_rental.merge(dim_customer[['customer_id','customer_key']], on='customer_id')
fact_rental = fact_rental.merge(dim_film[['film_id','film_key']], on='film_id')
fact_rental = fact_rental.merge(dim_store[['store_id','store_key']], on='store_id')
fact_rental = fact_rental.merge(dim_staff[['staff_id','staff_key']], on='staff_id')

# -------- FACT_PAYMENT --------
fact_payment = pd.read_sql("""
SELECT p.payment_date, p.amount,
r.customer_id,
r.staff_id,
i.film_id,
i.store_id
FROM payment p
JOIN rental r ON p.rental_id=r.rental_id
JOIN inventory i ON r.inventory_id=i.inventory_id
""", engine)

fact_payment['payment_date'] = pd.to_datetime(fact_payment['payment_date'])

#  date_key
fact_payment['date_key'] = fact_payment['payment_date'].dt.strftime('%Y%m%d').astype(int)

#  surrogate keys
fact_payment = fact_payment.merge(dim_customer[['customer_id','customer_key']], on='customer_id')
fact_payment = fact_payment.merge(dim_film[['film_id','film_key']], on='film_id')
fact_payment = fact_payment.merge(dim_store[['store_id','store_key']], on='store_id')
fact_payment = fact_payment.merge(dim_staff[['staff_id','staff_key']], on='staff_id')

# -------- FACT_INVENTORY --------
fact_inventory = pd.read_sql("""
SELECT store_id, film_id, COUNT(*) stock_count
FROM inventory
GROUP BY store_id, film_id
""", engine)

fact_inventory = fact_inventory.merge(dim_film[['film_id','film_key']], on='film_id')
fact_inventory = fact_inventory.merge(dim_store[['store_id','store_key']], on='store_id')

# ================================
# 4. ANALYSIS (باستخدام KEYS)
# ================================

# 1 Top Films
fact_rental.groupby('film_key').size().sort_values(ascending=False).head(10).plot(kind='barh', title="Top Films")
plt.show()

# 2 Top Customers
fact_rental.groupby('customer_key').size().sort_values(ascending=False).head(10).plot(kind='barh', title="Top Customers")
plt.show()

# 3 Late Rentals
fact_rental[fact_rental['is_late']==1].groupby('film_key').size().head(10).plot(kind='bar', title="Late Rentals")
plt.show()

# 4 Avg Duration
fact_rental.groupby('film_key')['actual_duration'].mean().head(10).plot(kind='bar', title="Avg Duration")
plt.show()

# 5 Rentals per Store
fact_rental.groupby('store_key').size().plot(kind='bar', title="Rentals per Store")
plt.show()

# 6 Rentals over Time
fact_rental.groupby('date_key').size().head(20).plot(title="Rentals Over Time")
plt.xticks(rotation=45)
plt.show()

# 7 Revenue per Customer
fact_payment.groupby('customer_key')['amount'].sum().sort_values(ascending=False).head(10).plot(kind='barh', title="Top Revenue Customers")
plt.show()

# 8 Revenue per Film
fact_payment.groupby('film_key')['amount'].sum().sort_values(ascending=False).head(10).plot(kind='barh', title="Revenue per Film")
plt.show()

# 9 Revenue over Time
fact_payment.groupby('date_key')['amount'].sum().head(20).plot(title="Revenue Over Time")
plt.xticks(rotation=45)
plt.show()

# 10 Revenue per Store
fact_payment.groupby('store_key')['amount'].sum().plot(kind='bar', title="Revenue per Store")
plt.show()

# 11 Inventory per Store
fact_inventory.groupby('store_key')['stock_count'].sum().plot(kind='bar', title="Inventory per Store")
plt.show()

# 12 Inventory per Film
fact_inventory.groupby('film_key')['stock_count'].sum().head(10).plot(kind='bar', title="Inventory per Film")
plt.show()

print("DONE ✅")
```
